```python
import pandas as pd
#leitura de dados
df = pd.read_csv ("https://pycourse.s3.amazonaws.com/temperature.csv")
#visualizando as primeiras 3 linhas
df.head(3)
```

```python
#tipos de dados
df.dtypes
```

```python
#transformando o tipo da coluna date (object) para datetime
df['date'] = pd.to_datetime(df['date'])
df.dtypes
```

```python
#setando o indice
df = df.set_index('date')
print(df.index)
```

```python
df.describe()
```

```python
#indexação por nome
#selecionando todas as linhas e coluna 1
df.loc[:, 'temperatura']
```

```python
#ordenando por uma coluna
df.sort_values(by='temperatura')
```

```python
#indexação por índice
#Selecionando todas as linhas e a coluna 1
df.iloc[:,1]
```
